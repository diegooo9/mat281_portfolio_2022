1. Lo primero será juntar toda la información en un _solo archivo_, para ello necesitamos seguir los siguientes pasos:

 * a) Crear el archivo **df_anio**, que contenga la información de **libertad_prensa_anio.csv** para cada año. Luego, normalice el nombre de las columnas a minúscula.
 * b) Encuentre y elimine el dato que esta duplicado en el archivo **df_codigo**.
 * c) Crear el archivo **df** que junte la información del archivo **df_anio** con **df_codigo** por la columna _codigo_iso_.
 
> **Hint**: Para juntar por _anio_ ocupe la función **pd.concat**. Para juntar información por columna ocupe **pd.merge**.

```python
# respuesta

df_anio = pd.DataFrame()
arreglo=[]

for name in archivos_anio:
    df_temp=pd.read_csv(name)
    arreglo.append(df_temp)
df_anio=pd.concat(arreglo, axis=0, sort=False)
M_columns=df_anio.columns
m_columns=[columna.lower() for columna in M_columns]
df_anio.columns=m_columns
paises=df_codigos["codigo_iso"].value_counts().keys().tolist()
index_paises=df_codigos["codigo_iso"].value_counts().tolist()
rep={}
rep2=[]
for i in range(0,len(index_paises)-1):
  if index_paises[i]>1:
    rep[paises[i]]=index_paises[i]
    rep2.append(paises[i])

df_codigos_f=df_codigos
for i in range(0,len(rep2)-1):
  for j in range(0,len(df_codigos["codigo_iso"]-1)):
    if df_codigos["codigo_iso"][j]==rep2[i]:
      if df_codigos["pais"][j]=="malo":
         df_codigos_f=df_codigos_f.drop(j)


df=pd.merge(df_codigos_f,df_anio,how="outer",on=["codigo_iso"])
print(df)






  
```

3. Desarrolle una función `resumen_df(df)` para encontrar el total de elementos distintos y vacíos por columnas.

```python
# respuesta

def resumen_df(Df):
    """
    funcion resumen con elementos distintos y vacios
    por columnas
    """
    nombres = Df.columns
    result = pd.DataFrame({'nombres': nombres})
    result['elementos_distintos']=0
    result['elementos vacios']=0
    for nombre in result.values():
       arreglo=[]
       for i in range(0,len(Df[nombre])):
         if Df[nombre][i] not in arreglo:
           arreglo.append(Df[nombre[i]])
           result['elementos_distintos']+=1
         else:
           continue
         R=isnull(Df[nombre[i]])
         if R==True:
           result['elementos vacios']+=1
         else:
           continue
    
    return result
  

```
